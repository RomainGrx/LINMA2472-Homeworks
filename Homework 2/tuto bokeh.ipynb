```python
import numpy as np
from bokeh.plotting import figure, output_file, show, output_notebook
```

```python
N = 4000
x = np.random.random(size=N)
y = np.random.random(size=N)
rad = np.random.random(size=N) / 40
colors = [
    "#%02x%02x%02x" % (int(r), int(g), 1) for r, g in zip(50+180*x, 30+180*y)
]

output_notebook()

TOOLS = "crosshair,pan,wheel_zoom,box_zoom,reset,box_select,lasso_select"

p = figure(tools=TOOLS)
p.circle(x, y, radius=rad, fill_color=colors, fill_alpha=.6,line_color=None)
show(p)
```

```python
import numpy as np
import pandas as pd
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.plotting import figure, output_notebook, show
from bokeh.sampledata.periodic_table import elements

N = 100
TITLE = "T_SNE of the Doc2Vec model"
TOOLS = "hover,pan,wheel_zoom,box_zoom,reset,save"
elements = dict(
    x=np.random.rand(N,1),
    y=np.random.rand(N,1),
    word=["prout"]*N
)

p = figure(title=TITLE,
           tools=TOOLS,
           toolbar_location="above",
           )
output_notebook()
p.background_fill_color = "#dddddd"
p.xaxis.axis_label = "t-SNE 1"
p.yaxis.axis_label = "t-SNE 2"
p.hover.tooltips = [
    ('word',"@word")
]

source = ColumnDataSource(elements)

p.circle("x","y",size=3,source=source,fill_alpha=.75)

show(p)
```
