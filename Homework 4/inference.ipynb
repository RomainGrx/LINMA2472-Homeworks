```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

plt.figure()
img = wgan.g(tf.random.normal((10,10)))[0]
plt.imshow(.5 * (img + 1))
```

```python
z = np.random.rand(1,10)
print(z)
plt.figure()
fig, axis = plt.subplots(5, 5, figsize=(10,10))
axis = np.array(axis).flatten()
for x in range(1,6):
    for y in range(1,6):
        z[0,8] = 5/x
        z[0,9] = 5/x
        img = .5*(1+wgan.g(z)[0])
        axis[5*(x-1)+y-1].imshow(img)
```
